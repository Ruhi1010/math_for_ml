```python
# 1. Basic Arithmetic Operations
my_value = 2 * (3 + 2)**2 / 5 - 4
print(my_value) # prints 6.0
```

```python
# 2. Function Definition and Usage
my_value = 2 * ((3 + 2)**2 / 5) - 4
print(my_value) # prints 6.0
```

```python
# 3. Function Definition and Evaluation
def f(x):
    return 2 * x + 1
x_values = [0, 1, 2, 3]
for x in x_values:
    y = f(x)
    print(y)
```

```python
# 4. Plotting a Function
from sympy import *
x = symbols('x')
f = 2*x + 1
plot(f)
```

```python
# 5. Plotting a Quadratic Function
from sympy import *
x = symbols('x')
f = x**2 + 1
plot(f)
```

```python
# 6. Plotting a 3D Function
from sympy import *
from sympy.plotting import plot3d
x, y = symbols('x y')
f = 2*x + 3*y
plot3d(f)
```

```python
from sympy import *
i,n = symbols('i n')
# iterate each element i from 1 to n,
# then multiply and sum
summation = Sum(2*i,(i,1,n))
# specify n as 5,
# iterating the numbers 1 through 5
up_to_5 = summation.subs(n, 5)
print(up_to_5.doit())
```

```python
from sympy import *
x = symbols('x')
expr = x**2 / x**5
print(expr.simplify())
```
